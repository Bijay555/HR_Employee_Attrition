```python
import pandas as pd
pd.options.display.max_rows=10
pd.options.display.max_columns=50
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('data.csv')
df.info()
```

```python
df.head()
```

```python
# Doing one hot encoding of categorical variables
df['BusinessTravel']=df['BusinessTravel'].replace(['Non-Travel','Travel_Rarely','Travel_Frequently'], [0,1,2]) 
# Departments
df['Department']=df['Department'].astype('category')
dummies=pd.get_dummies(df['Department'],prefix='Department')
df = pd.concat([df, dummies], axis=1)
del df['Department']
# Education Field
df['EducationField']=df['EducationField'].astype('category')
dummies=pd.get_dummies(df['EducationField'],prefix='EducationField')
df = pd.concat([df, dummies], axis=1)
del df['EducationField']
# Gender
df['Gender']=df['Gender'].astype('category')
dummies=pd.get_dummies(df['Gender'],prefix='Gender')
df = pd.concat([df, dummies], axis=1)
del df['Gender']
# JobRole
df['JobRole']=df['JobRole'].astype('category')
dummies=pd.get_dummies(df['JobRole'],prefix='JobRole')
df = pd.concat([df, dummies], axis=1)
del df['JobRole']
# MaritalStatus
df['MaritalStatus']=df['MaritalStatus'].astype('category')
dummies=pd.get_dummies(df['MaritalStatus'],prefix='MaritalStatus')
df = pd.concat([df, dummies], axis=1)
del df['MaritalStatus']
# Dropping over 18 as its giving no data
del df['Over18']

df['OverTime']=df['OverTime'].replace(['No','Yes'], [0,1]) 

df.head()
```

```python
df.info()
```

```python
# One hot encoding Attrition column
# Yes -> 1
# No -> 0

df['Attrition']=df['Attrition'].replace(['Yes', 'No'], [1,0]) 
df['Attrition']=df['Attrition'].astype('category')
```

```python
# Writing the main data set in a csv format
df.to_csv('MainEncodedData.csv')
```

```python
df.shape
```

```python
(df['Attrition'].value_counts()/1470)*100
```

```python
y=df['Attrition']
x=df.drop(df.columns[1], axis = 1)
```

Ok so now i know my data is biased towards non attrition
### Fixing this using smote

```python
df.shape
```

```python
sm = SMOTE(random_state=789, ratio =1,k_neighbors=9,kind='svm')
```

```python
sdf=sm.fit_sample(x,y)
```

```python
d=pd.DataFrame(list(sdf[0]),columns=['Age', 'BusinessTravel', 'DailyRate', 'DistanceFromHome',
       'Education', 'EmployeeCount', 'EmployeeNumber',
       'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement', 'JobLevel',
       'JobSatisfaction', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked',
       'OverTime', 'PercentSalaryHike', 'PerformanceRating',
       'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel',
       'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager', 'Department_Human Resources',
       'Department_Research & Development', 'Department_Sales',
       'EducationField_Human Resources', 'EducationField_Life Sciences',
       'EducationField_Marketing', 'EducationField_Medical',
       'EducationField_Other', 'EducationField_Technical Degree',
       'Gender_Female', 'Gender_Male', 'JobRole_Healthcare Representative',
       'JobRole_Human Resources', 'JobRole_Laboratory Technician',
       'JobRole_Manager', 'JobRole_Manufacturing Director',
       'JobRole_Research Director', 'JobRole_Research Scientist',
       'JobRole_Sales Executive', 'JobRole_Sales Representative',
       'MaritalStatus_Divorced', 'MaritalStatus_Married',
       'MaritalStatus_Single'])
d_y=pd.DataFrame(list(sdf[1]),columns=['Attrition'])
d=pd.concat([d,d_y],axis=1)
d.shape
```

```python
d.head()
```

```python
(d['Attrition'].value_counts()/2465)*100
```

```python
#Saving smote attemp data
# Writing the main data set in a csv format
d.to_csv('SmoteData_1.csv')
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=786)
X_resampled, y_resampled = ros.fit_sample(x, y)
```

```python
d2=pd.DataFrame(X_resampled,columns=['Age', 'BusinessTravel', 'DailyRate', 'DistanceFromHome',
       'Education', 'EmployeeCount', 'EmployeeNumber',
       'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement', 'JobLevel',
       'JobSatisfaction', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked',
       'OverTime', 'PercentSalaryHike', 'PerformanceRating',
       'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel',
       'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager', 'Department_Human Resources',
       'Department_Research & Development', 'Department_Sales',
       'EducationField_Human Resources', 'EducationField_Life Sciences',
       'EducationField_Marketing', 'EducationField_Medical',
       'EducationField_Other', 'EducationField_Technical Degree',
       'Gender_Female', 'Gender_Male', 'JobRole_Healthcare Representative',
       'JobRole_Human Resources', 'JobRole_Laboratory Technician',
       'JobRole_Manager', 'JobRole_Manufacturing Director',
       'JobRole_Research Director', 'JobRole_Research Scientist',
       'JobRole_Sales Executive', 'JobRole_Sales Representative',
       'MaritalStatus_Divorced', 'MaritalStatus_Married',
       'MaritalStatus_Single'])
d2_y=pd.DataFrame(y_resampled,columns=['Attrition'])
d2=pd.concat([d2,d2_y],axis=1)
d2.shape
```
